```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体为 SimHei
matplotlib.rcParams['axes.unicode_minus'] = False  # 正确显示负号

def plot_year_count(df):
    year_counts = df['年份'].value_counts().sort_index()
    
    plt.figure(figsize=(10, 6))  # 设置图形的尺寸
    plt.bar(year_counts.index, year_counts.values, color='blue')  # 创建柱状图
    plt.xlabel('年份')  # 设置 x 轴标签
    plt.ylabel('数量')  # 设置 y 轴标签
    plt.title('每个年份的报告数量')  # 设置图形标题
    plt.xticks(rotation=45)  # 旋转 x 轴标签以便阅读
    plt.grid(True)  # 显示网格
    plt.show()  # 显示图形
```

```python
df_600 = pd.read_excel('c上交所主板_年报链接.xlsx')
df_600['年份'] = df_600['报告名称'].str.extract(r'(\d{4})年')

print(df_600[['报告名称', '年份']])

plot_year_count(df_600)
```

```python
df_688 = pd.read_excel('c上交所科创板_年报链接.xlsx')
df_688['年份'] = df_688['报告名称'].str.extract(r'(\d{4})年')

print(df_688[['报告名称', '年份']])

plot_year_count(df_688)
```

```python
df_000 = pd.read_excel('c深交所主板_年报链接.xlsx')
df_000['年份'] = df_000['报告名称'].str.extract(r'(\d{4})年')

print(df_000[['报告名称', '年份']])

plot_year_count(df_000)
```

```python
df_300 = pd.read_excel('c深交所创业板_年报链接.xlsx')
df_300['年份'] = df_300['报告名称'].str.extract(r'(\d{4})年')

print(df_300[['报告名称', '年份']])

plot_year_count(df_300)
```

```python
df_600['板块'] = '上交所主板'
df_688['板块'] = '上交所科创板'
df_000['板块'] = '深交所主板'
df_300['板块'] = '深交所创业板'
```

```python
df = pd.concat([df_600, df_688, df_000, df_300])
print(df.shape)
print(df.head())
```

```python
def plot_year_count_by_sector(df):
    data = df.pivot_table(index='年份', columns='板块', aggfunc='size', fill_value=0)
    
    plt.figure(figsize=(10, 5))
    data.plot(kind='bar', stacked=True, color=['red', 'green', 'blue', 'yellow'])
    
    plt.xlabel('年份')
    plt.ylabel('数量')
    plt.title('各板块每个年份的报告数量')
    
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.legend(title='板块', loc='upper left', prop={'family': 'SimHei'})
    plt.savefig('年报数量统计.png', dpi=600)
    plt.show()
    
plot_year_count_by_sector(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def plot_year_count_by_sector_seaborn(df):
    data = df.pivot_table(index='年份', columns='板块', aggfunc='size', fill_value=0)
    
    plt.figure(figsize=(10, 5))
    
    palette = sns.color_palette("pastel")  
    
    bottom = np.zeros(len(data))
    
    for col in data.columns:
        plt.bar(data.index, data[col], bottom=bottom, label=col, color=palette.pop(0))
        bottom += data[col].values
    
    plt.xlabel('年份')
    plt.ylabel('数量')
    plt.title('各板块每个年份的报告数量')
    plt.xticks(rotation=45)
    plt.grid(True)
    
    plt.legend(title='板块', loc='upper left', prop={'family': 'SimHei'})
    plt.savefig('年报数量统计_seaborn.png', dpi=600)
    plt.show()

plot_year_count_by_sector_seaborn(df)
```

```python
def plot_year_count_by_sector_seaborn(df):
    data = df.pivot_table(index='年份', columns='板块', aggfunc='size', fill_value=0)
    
    plt.figure(figsize=(10, 5))
    palette = sns.color_palette("pastel")
    hatches = ['//', '||', '+', 'xx']  
    bottom = np.zeros(len(data))
    
    for i, col in enumerate(data.columns):
        plt.bar(data.index, data[col], bottom=bottom, label=col, color=palette[i], hatch=hatches[i % len(hatches)])
        bottom += data[col].values
    
    plt.xlabel('年份')
    plt.ylabel('数量')
    plt.title('各板块每个年份的报告数量')
    plt.xticks(rotation=45)
    plt.grid(True)
    
    plt.legend(title='板块', loc='upper left', prop={'family': 'SimHei'})
    plt.savefig('年报数量统计_seaborn_hatch.png', dpi=600)
    plt.show()

plot_year_count_by_sector_seaborn(df)
```
